### 1.5 Data cleaning: 'followers' column

###### Special list to clean

```python
followers = []

for i in df.index:
    row = df.iloc[i, -1]
    row = row.replace("[", "").replace("]","").replace("'","").replace(" ","").split(",")
    followers.append(row)

len(followers)
```

```python
# Hundred of thousands
n = 0

for lista in followers:
    x = 0
    for i in lista:
        if len(i) == 4:
            followers[n][x] = (i.replace("k", "000"))
            x += 1
        else:
            x +=1
    n += 1    
    
followers
```

```python
# Thousands
n = 0

for lista in followers:
    x = 0
    for i in lista:
        followers[n][x] = (i.replace("k", "00"))
        x += 1
    n += 1    
    
followers
```

```python
# Millions
n = 0

for lista in followers:
    x = 0
    for i in lista:
        if len(i) <= 3:
            followers[n][x] = (i.replace("m", "000000"))
            x += 1
        else:
            x +=1
    n += 1    
    
followers
```

```python
df.iloc[30,-1]
```

```python
# Only case with more than one hundred million followers: Kevin Hart
followers[30][2] = "128000000"
```

```python
# Cast everything to integer
n = 0

for lista in followers:
    x = 0
    for i in lista:
        try:
            followers[n][x] = int(i)
            x += 1
        except:
            followers[n][x] = 0
            x += 1
    n += 1    
    
followers
```

###### Build the column

```python
df["followers"] = [sum(row) for row in followers]
df["followers"]
```

## 2. Feature engineering 

Build new columns with the existing columns

### 2.1 Year column

```python
# Column 'Year'
df['year'] = df.premiere.dt.year
df['year']
```

### 2.2 Month column

```python
# Column 'Month'
df['month'] = df.premiere.dt.month

# Convert the results to string
df['month'] = df['month'].replace(1, 'january')
df['month'] = df['month'].replace(2, 'february')
df['month'] = df['month'].replace(3, 'march')
df['month'] = df['month'].replace(4, 'april')
df['month'] = df['month'].replace(5, 'may')
df['month'] = df['month'].replace(6, 'june')
df['month'] = df['month'].replace(7, 'july')
df['month'] = df['month'].replace(8, 'august')
df['month'] = df['month'].replace(9, 'september')
df['month'] = df['month'].replace(10, 'october')
df['month'] = df['month'].replace(11, 'november')
df['month'] = df['month'].replace(12, 'december')

df['month']
```

### 2.3 Year/Month column

```python
# Column 'year/month': will allow group films according to their year and month of release.
df['year/month'] = df.premiere.apply(lambda dt: dt.replace(day=1))
df['year/month']
```

### 2.4 Weekday column

```python
# Column 'weekday'
df['weekday'] = df.premiere.dt.weekday

# Convert the results to string
df['weekday'] = df['weekday'].replace(0, 'Monday')
df['weekday'] = df['weekday'].replace(1, 'Tuesday')
df['weekday'] = df['weekday'].replace(2, 'Wednesday')
df['weekday'] = df['weekday'].replace(3, 'Thursday')
df['weekday'] = df['weekday'].replace(4, 'Friday')
df['weekday'] = df['weekday'].replace(5, 'Saturday')
df['weekday'] = df['weekday'].replace(6, 'Sunday')

# Check the column
df['weekday']
```

### 2.5. Column "IMDB Score rolling average": rating's rolling average

```python
# Sort DataFrame by "premiere"
df.sort_values(by = ["premiere"], inplace = True)
df.reset_index(drop = True, inplace = True)
```

```python
df['score_rolling_average'] = df['imdb_score'].rolling(10).mean()
df['score_rolling_average']
```
